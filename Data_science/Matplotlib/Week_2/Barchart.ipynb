```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2
xvar = range(len(linear_data))
plt.bar(xvar,linear_data, width=0.3)
new_xvals = [items+0.3 for items in xvar]
plt.bar(new_xvals,quadratic_data, width=0.3, color ='red')
```

```python
from random import randint
linear_err = [randint(0,15) for item in linear_data]
plt.bar(xvar, linear_data, width=0.3, yerr = linear_err)
```

```python
plt.figure()
plt.bar(xvar,linear_data, width=0.3, color='b')
plt.bar(xvar,quadratic_data, width=0.3, bottom= linear_data, color='r')
```

```python
plt.figure()

plt.barh(xvar,linear_data, height=0.3, color='b')
plt.barh(xvar,quadratic_data, height=0.3, left = linear_data, color='r')
```
