## DataFrame.plot

```python
np.random.seed(223)
df = pd.DataFrame({'A': np.random.rand(365).cumsum(0),
                   'B': np.random.rand(365).cumsum(0)+20,
                   'C': np.random.rand(365).cumsum(0)-20},
                     index = pd.date_range('1/1/2020', periods=365))
df.head()
```

```python
df.plot();
```

```python
df.plot('A','B', kind = 'scatter')
```

```python
df.plot.scatter('A','C', c='B', s=df['B'], colormap='viridis')
```

```python
ax = df.plot.scatter('A','C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')
```

```python
df.plot.box()
```

```python
df.plot.hist(alpha=0.7);
```

```python
df.plot.kde();
```
